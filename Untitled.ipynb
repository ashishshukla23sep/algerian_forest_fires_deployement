```python
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    return lwr_bound,upr_bound
lwr,upper=detect_outliers_iqr(df['Rain'])
print("lower Bound:",lwr,"Upper Bound:",upper)
print(len(df[(df['Rain']>lwr)&(df['Rain']<upper)]))
print(df[(df['Rain']>lwr)&(df['Rain']<upper)]['Classes'].value_counts())
```

### Model Evaluation function

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
model_evel={}
def model_evalution(x,y,models,sss=False,Stdsclr=False):
    if Stdsclr:
        if sss:
            sss=StratifiedShuffleSplit(n_splits=1,random_state=10,test_size=0.2)
            train_index,test_index = next(sss.split(df.drop(columns='Classes'),df['Classes']))
            x_train,x_test=x.iloc[train_index],x.iloc[test_index]
            y_train,y_test=y.iloc[train_index],y.iloc[test_index]
            x_train=scalar.fit_transform(x_train)
            x_test=scalar.transform(x_test)
        else:
            x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
            x_train=scalar.fit_transform(x_train)
            x_test=scalar.transform(x_test)
    else:
        if sss:
            sss=StratifiedShuffleSplit(n_splits=1,random_state=10,test_size=0.2)
            train_index,test_index = next(sss.split(df.drop(columns='Classes'),df['Classes']))
            x_train,x_test=x.iloc[train_index],x.iloc[test_index]
            y_train,y_test=y.iloc[train_index],y.iloc[test_index]
            
        else:
            x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
            
    for model_name,model in models.items():
        score={}
        model.fit(x_train,y_train)
        y_pred=model.predict(x_test)
        score["mse_score"]=mean_squared_error(y_test,y_pred)
        score["r2_score_test"]=r2_score(y_test,y_pred)
        score["r2_score_train"]=model.score(x_train,y_train)
        score["Croos_val_score"]=cross_val_score(model,x,y,cv=5,scoring='r2').mean()
        model_evel[model_name]=score
    return model_evel
```
